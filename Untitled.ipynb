```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import html5lib
import numpy as np
```

```python
wikiurl = "http://en.wikipedia.org/wiki/List_of_cities_in_Pennsylvania"
response = requests.get(wikiurl)
print(response.status_code)
```

```python
soup = bs(response.text, 'html.parser')
pennTable = soup.find('table', {'class':"wikitable"})
pennTable
```

```python
df = pd.read_html(str(pennTable), flavor="bs4")
df = pd.DataFrame(df[0])
df.head()
```

```python
df.columns = ['Name', 'Type', 'County', 'Class', 'Population (2018 est.)', 'Incorporation year', 'Sq Miles']
df.head()
```

```python
df = df.drop(columns=['Type', 'Class'])
df.head()
```
